# Series

```python
import pandas as pd

a = pd.Series([1, 2, 3, 4, 5])  # default index
print(a)
```

```python
import pandas as pd # useful for date and time manipulation

a = pd.Series([1, 2, 3, 4, 5],index = ['a', 'b', 'c', 'd', 'e'])  # index can be changed
print(a)

print([a['d']]) # index can be used to access elements

print(a['a':'c']) # slicing is possible

print(a[[0, 2, 4]]) # indexing is possible
```

```python
a = pd.Series([1,2,3,4,5])  # arithmetic operations are possible
print(a+21)

print(a-2)

print(a*2)

print(a/2)
```

```python
a=pd.Series([1,2,3,4,5])
b=pd.Series([10,20,30,40,50])

print(a+b)
```

```python
a = pd.Series([1,2,3,4,5])

print(a>3)

print(a<3)

print(a==3)

print(a!=3)
```

```python
a = pd.Series([1,2,3,4,5])

print(a.apply(lambda x: x*2)) # lambda function is used to apply a function to all elements of a series
```

```python
a = pd.Series([1,2,3,4,5])
print(a.apply(lambda x:'even' if x%2==0 else 'odd'))
```

```python
#mathematical functions

print(a.sum())

print(a.mean()) # formula is mean = sum/n

print(a.max())

print(a.min())

print(a.median()) # formula is median = (n+1)/2

print(a.std())  # formula is standard deviation = sqrt(sum((x - mean)^2) / n)

print(a.var()) # formula is variance = sum((x - mean)^2) / n

```

```python
a=pd.Series([1,2,2,4,5,6,1])
print(a.describe())
print("_______________________________","\n")


print(a.unique())
print("_______________________________","\n")


print(a.nunique()) # number of unique elements
print("________________________________","\n")


print(a.value_counts()) # frequency of each element
print("________________________________","\n")


print(a.is_unique)
print("________________________________","\n")


print(a.sort_values())
print("________________________________","\n")


print(a.sort_index())
print("________________________________","\n")


print(a.cumsum()) # meaning of cumulative sum is sum of first n elements
```
