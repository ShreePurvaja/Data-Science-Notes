```python
#concat and merge
import pandas as pd
data1 = {
  'ID': ['1','2','3','4','5'],
  'Feature1': ['A','B','C','D','E'],
  'Feature2': ['L','M','N',"O",'P']
}



df1 = pd.DataFrame(data1, columns = ["ID", "Feature1", "Feature2"])


d2 = {
  'ID': ['1','3','6','7','8'],
  'Feature3': ['AB','BC','CD','DE','EF'],
  'Feature4': ['KL','LM','M-N',"N-O",'O-P']
}


df2 = pd.DataFrame(d2, columns = ["ID", "Feature3", "Feature4"])
print(df1)
print(df2)
```

```python
display(pd.concat([df1,df2],axis=0)) #axis 0 means row and axis 1 means column
```

```python
#append - doen't work in dataframes
display(df1.append(df2))
```

```python
#merge

display(pd.merge(df1,df2,on="ID",how = "right"))
```

```python
display(pd.merge(df1,df2,on="ID",how = "inner"))
```

```python
display(pd.merge(df1,df2,on="ID",how = "left"))
```

```python
display(pd.merge(df1,df2,on="ID",how = "outer"))
```

```python
display(df1.merge(df2,how = "cross")) #cross join
```

```python
display(pd.merge(df1,df2,on="ID",how = "outer",indicator = True)) # merge works by giving a common column
```

```python
#simple method
a=df1.merge(df2,on="ID",how = "outer",indicator = True) # merge works by giving a common column
a[a["_merge"] != "both"]
```

```python
# method - 1
display(pd.merge(df1,df2,on="ID",how="outer", indicator=True).query('_merge=="left_only" or _merge=="right_only"'))
```

```python
#joins can be used in DF
display(df1.join  (df2, lsuffix="_left",rsuffix="_right"))
```

```python
display(df1.join(df2["Feature3"],lsuffix="_left",rsuffix="_right"))
```

```python
display(df1.join(df2["Feature3"],lsuffix="_left",rsuffix="_right").join(df2["Feature4"],lsuffix="_left",rsuffix="_right"))
```

```python
# groupby
import pandas as pd
nd=pd.read_csv("https://raw.githubusercontent.com/nethajinirmal13/Training-datasets/refs/heads/main/titanic.csv")
display(nd)
```

```python
a = nd.groupby("Survived")
a.apply(display) # applying display in the group by
```

```python
a.size()
```

```python
a=nd.groupby(["Survived","Sex","Embarked"]).size()
a
```

```python
#find the no of crew member
crew=nd[nd["Fare"]==0].count()
display(crew)
```

```python
a=nd.groupby("Survived").agg({"Age":["mean","max","min"]})
display(a)
```

```python
#crew member count
a=nd.groupby("Survived").agg({"Age":["mean","max","min"]})
display(a)
```
