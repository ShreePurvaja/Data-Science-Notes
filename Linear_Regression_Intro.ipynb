```python
import numpy as np
x=[1,2,4,3,5]
y=[1,3,3,2,5]
mean_x = np.mean(x)
mean_y = np.mean(y)




def linreg(x,y):
# Using the formula to calculate b1 and b2
  numer = 0
  denom = 0
  m=len(x)
  for i in range(m):
    numer += (x[i] - mean_x) * (y[i] - mean_y)
    denom += (x[i] - mean_x) ** 2



  M = numer / denom




  c = mean_y - (M*mean_x)
  print(M,c)
  return M,c
a=linreg(x,y)
print(a)
b=[]
for i in x:
  b.append((a[0]*i+a[1]))
print(b)
```

```python
import matplotlib.pyplot as plt   #plot only come for simple datas
# Plot the actual points
plt.scatter(x, y, color='blue', label='Actual Points')
# Plot the linear regression line
plt.plot(x, b, color='red', label='Linear Regression')
# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Linear Regression Plot')
plt.legend()
# Display the plot
plt.show()
```

```python
import pandas as pd
data = pd.read_csv("https://raw.githubusercontent.com/nethajinirmal13/Training-datasets/main/1.01.%20Simple%20linear%20regression.csv")
display(data)

```

```python
from sklearn.linear_model import LinearRegression
X = data.drop('GPA', axis=1)    #to specify the gpa column    x is in future ,future is only 2D format
y = data['GPA']
model = LinearRegression()    #sat vechu gpa find pandrom
model.fit(X, y)
print(model.coef_)  #m value
print(model.intercept_) #c value
print(model.predict([[3740]]))    #consider input is given as two dimentional array
```
