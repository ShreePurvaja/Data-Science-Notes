```python
#stacked bar chart
from matplotlib import pyplot as plt
import numpy as np



countries = ['USA', 'India', 'China', 'Russia', 'Germany']
bronzes = np.array([38, 17, 26, 19, 15])
silvers = np.array([37, 23, 18, 18, 10])
golds = np.array([46, 27, 26, 19, 17])



plt.bar(countries, golds, width=0.5, label='golds', color='gold', bottom=silvers+bronzes) #saying that below gold solver and gold should come
plt.bar(countries, silvers, width=0.5, label='silvers', color='silver', bottom=bronzes) #it will not show any error but it will get over stacked
plt.bar(countries, bronzes, width=0.5, label='bronzes', color='#CD853F')




plt.ylabel("Medals")
plt.xlabel("Countries")
plt.legend(loc="upper right")
plt.title("2019 Olympics Top Scorers")
```

```python
#Task of the day

# 1. want only 1930 yr data
# 2. x axis country
# 3. y axis total medal bar (no stacked bar)

import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/nethajinirmal13/Training-datasets/main/COMMONWEALTH%20GAMES%201930-2018%20%20-%20MEDAL%20TALLY.csv")
df
```

```python
df=df[df["Year"]==1930]
df
```

```python
year_1930_data = df[df['Year'] == 1930]
countries = year_1930_data['Country']
# total_medals = year_1930_data['Total']




year_1930_data['Total'] = pd.to_numeric(year_1930_data['Total'], errors='coerce') #'errors = coerce 'if there is any error it automatically puts zero there




plt.figure(figsize=(12, 6))
plt.bar(countries, year_1930_data['Total'])
plt.xlabel("Country")
plt.ylabel("Total Medals")
plt.title("Commonwealth Games 2018 Medal ")
plt.xticks(rotation=45, ha='right') #if not tilted the labels get overlapsed
plt.tight_layout()
plt.show()
```

```python
# prompt: write a plot  that represents total number of restaurants for each cuisines
import matplotlib.pyplot as plt
import pandas as pd
# Load the dataset
data = pd.read_csv("https://raw.githubusercontent.com/nethajinirmal13/Training-datasets/refs/heads/main/zomato/zomato.csv")
# Count the number of restaurants for each cuisine
cuisine_counts = data['Cuisines'].value_counts()
# Create the bar plot
plt.figure(figsize=(12, 6))
plt.bar(cuisine_counts.index[:30], cuisine_counts.values[:30], color='skyblue')  # Plotting only the top 20 cuisines for better visualization
plt.xlabel('Cuisines')
plt.ylabel('Number of Restaurants')
plt.title('Total Number of Restaurants for Each Cuisine (Top 20)')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```

```python
# prompt: write a plot that represents total number of restaurants in each locality with 30 datas

import matplotlib.pyplot as plt
import pandas as pd

# Load the dataset
data = pd.read_csv("https://raw.githubusercontent.com/nethajinirmal13/Training-datasets/refs/heads/main/zomato/zomato.csv")

# Count the number of restaurants in each locality
locality_counts = data['Locality'].value_counts()

# Create the bar plot for the top 30 localities
plt.figure(figsize=(12, 6))
plt.bar(locality_counts.index[:30], locality_counts.values[:30], color='skyblue')
plt.xlabel('Locality')
plt.ylabel('Number of Restaurants')
plt.title('Total Number of Restaurants in Each Locality (Top 30)')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```

```python
# prompt: write a plot that represents top 20 restaurants based on votes

# Load the dataset
data = pd.read_csv("https://raw.githubusercontent.com/nethajinirmal13/Training-datasets/refs/heads/main/zomato/zomato.csv")

# Assuming 'Votes' column represents the votes for each restaurant
# Sort the restaurants based on votes in descending order and get the top 20
top_restaurants = data.sort_values(by='Votes', ascending=False).head(20)

# Create the bar plot
plt.figure(figsize=(12, 6))
plt.bar(top_restaurants['Restaurant Name'], top_restaurants['Votes'], color='skyblue')
plt.xlabel('Restaurant Name')
plt.ylabel('Number of Votes')
plt.title('Top 20 Restaurants Based on Votes')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```

```python
# prompt: write a plot that represents the 30 different cuisines based on ratings from 0 to 5

import matplotlib.pyplot as plt
import pandas as pd

# Load the dataset
data = pd.read_csv("https://raw.githubusercontent.com/nethajinirmal13/Training-datasets/refs/heads/main/zomato/zomato.csv")

# Assuming 'Aggregate rating' represents the ratings
# Group by cuisine and calculate the average rating
cuisine_ratings = data.groupby('Cuisines')['Aggregate rating'].mean()

# Create the bar plot for the top 30 cuisines
plt.figure(figsize=(12, 6))
plt.bar(cuisine_ratings.index[:30], cuisine_ratings.values[:30], color='skyblue')
plt.xlabel('Cuisines')
plt.ylabel('Average Rating')
plt.title('Average Rating for Each Cuisine (Top 30)')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```

```python
# prompt: write a piechart for has online delivery data


import pandas as pd
import matplotlib.pyplot as plt


# Load the data
data = pd.read_csv("https://raw.githubusercontent.com/nethajinirmal13/Training-datasets/refs/heads/main/zomato/zomato.csv")


# Check for missing values in the 'Has Online delivery' column
print(data['Has Online delivery'].isnull().sum())


# Assuming 'Has Online delivery' has values like 'Yes' and 'No'
# Count the occurrences of 'Yes' and 'No' in the 'Has Online delivery' column
online_delivery_counts = data['Has Online delivery'].value_counts()


# Create the labels for the pie chart
labels = online_delivery_counts.index


# Create the sizes for the pie chart
sizes = online_delivery_counts.values


# Create the pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)


# Add a title (optional)
plt.title('Restaurants with Online Delivery')


# Ensure the circle's proportion
plt.axis('equal')


# Display the chart
plt.show()
```
