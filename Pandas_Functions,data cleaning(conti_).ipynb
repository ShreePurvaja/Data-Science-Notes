```python
import pandas as pd
airbnb = pd.read_csv('https://raw.githubusercontent.com/nethajinirmal13/Training-datasets/main/airbnb.csv?raw=true', index_col = 'Unnamed: 0')
display(airbnb)
airbnb.shape
```

```python
airbnb.info()#check the data type of each column
```

```python
airbnb["price"]=airbnb["price"].str.replace("$","") #remove $
airbnb["price"]=airbnb["price"].astype(float) #convert string to float
```

```python
display(airbnb) #now the $ symbol in price is removed
```

```python
airbnb.isnull().sum() #check for null values
```

```python
airbnb["coordinates"]=airbnb["coordinates"].str.replace("(","")
airbnb["coordinates"]=airbnb["coordinates"].str.replace(")","")
display(airbnb) #removed () in coordinates
```

```python
a=airbnb["coordinates"].str.split(",",expand=True) #splits into different columns
display(a)                                         #expand is used to create columns instead of displaying in list
```

```python
airbnb["lattitude"]=a[0]
airbnb["longitude"]=a[1]
display(airbnb)
```

```python
airbnb["lattitude"] = airbnb["lattitude"].astype(float) #type conversion
airbnb["longitude"] = airbnb["longitude"].astype(float)
display(airbnb)
```

```python
airbnb.drop(columns="coordinates", inplace=True)#deletes the coordinates colun
display(airbnb)
```

```python
airbnb["listing_id"].nunique() #unique values
```

```python
b=airbnb["listing_id"].duplicated(keep=False) #duplicate values
airbnb[b].sort_values("listing_id") #keep = False means shows all duplicated values
```

```python
b = airbnb[airbnb["listing_id"].duplicated()] #
display(b.index)
```

```python
b=airbnb["listing_id"].duplicated(keep='first') #duplicate values
airbnb[b].sort_values("listing_id") #keep = False means shows all duplicated values
print(airbnb[b].sort_values("listing_id").shape)
```

```python
airbnb = airbnb.drop_duplicates(subset="listing_id", keep="first") # removes the duplicates
display(airbnb)
```

```python
#duplicate values after removing
b = airbnb["listing_id"].duplicated(keep = 'first')
airbnb[b].sort_values("listing_id")
print(airbnb[b].sort_values("listing_id").shape)
```

```python
airbnb["room_type"].unique() #unique rooms
```

```python
#Cleaning data for room_type... replacing the repeated values
airbnb["room_type"]=airbnb["room_type"].str.replace("Entire home/apt","home")
airbnb["room_type"]=airbnb["room_type"].str.replace("private room","private")
airbnb["room_type"]=airbnb["room_type"].str.lower() #lowercase
airbnb["room_type"]=airbnb["room_type"].str.strip() #removes spaces
```

```python
airbnb["room_type"].unique()
```

```python
airbnb.groupby("room_type").agg({"price":["mean","max","min"]})
```

```python
# Impute price based on conditions
airbnb.loc[(airbnb['price'].isna()) & (airbnb['room_type'] == 'home'), 'price'] = 208.7 #wherever the price is null and room_type is home, replace it with 208.7
airbnb.loc[(airbnb['price'].isna()) & (airbnb['room_type'] == 'private'), 'price'] = 89.02 #wherever the price is null and room_type is private, replace it with 89.02
airbnb.loc[(airbnb['price'].isna()) & (airbnb['room_type'] == 'shared room'), 'price'] = 72.41 #wherever the price is null and room_type is shared room, replace it with 72.41
```
