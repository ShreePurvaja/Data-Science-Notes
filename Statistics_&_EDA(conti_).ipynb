```python
#IQR using Numpy

import numpy as np

data = [50, 60, 70, 80, 90, 100, 110, 120, 130, 140]


Q1 = np.percentile(data, 25)
# Third quartile (Q3)
Q3 = np.percentile(data, 75)
# Interquaritle range (IQR)
IQR = Q3 - Q1
print(IQR)
print(Q1,Q3)
```

```python
import matplotlib.pyplot as plt

import pandas as pd

data = pd.read_csv('/content/diabetes.csv')

data["BloodPressure"].hist(bins = 20)
```

```python
#finding the boundary values

print("Highest allowed",data['BloodPressure'].mean() + 3*data['BloodPressure'].std())
print("Lowest allowed",data['BloodPressure'].mean() - 3*data['BloodPressure'].std())
```

```python
#trimming the unnecessary data to handle outliers
b = data[(data['BloodPressure'] < 127.1) & (data['BloodPressure'] > 11)]
display(b)
```

```python
#capping / Imputation
import numpy as np

data['BloodPressure']=np.where(data['BloodPressure']>127.1,127.1,data['BloodPressure']) # where function is similar to if
data['BloodPressure']=np.where(data['BloodPressure']<11.03,11.03,data['BloodPressure'])
display(data['BloodPressure'])
```

```python
#apply capping - nested function
import numpy as np
df = pd.read_csv('/content/diabetes.csv')

df['BloodPressure'] = np.where(
   df['BloodPressure']>127.1,
   127.1,
   np.where(
       df['BloodPressure']<11.03,
       11.03,
       df['BloodPressure']
   )
)
df.boxplot("BloodPressure")
```

```python
#Detetion tech
#Finding IQR
import pandas as pd

data=pd.read_csv('/content/diabetes.csv')

percentile25=data["BloodPressure"].quantile(0.25)
precentile75 = data["BloodPressure"].quantile(0.75)
IQR = precentile75 - percentile25
print(IQR)
print(percentile25)
print(precentile75)
```

```python
#Finding the upper and lower limit
upper_limit=precentile75 + 1.5*IQR
lower_limit=percentile25 - 1.5*IQR
print(upper_limit)
print(lower_limit)
```
