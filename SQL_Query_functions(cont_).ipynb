```python
mycursor.execute("select sum(bmi)/count(bmi) from guvi1.Diabetes " )  # fetch the average bmi by using formula
out=mycursor.fetchall()   # fetching all the data
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))   # psql is like sql
```

```python
mycursor.execute("select count(*) from guvi1.Diabetes where (bmi >= 19 and bmi <= 25) and outcome = 1  " )  # fetch the count of diabetic patients whose bmi is between 19 and 25
out=mycursor.fetchall()   # fetching all the data
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))   # psql is like sql
```

```python
mycursor.execute("select count(*) from guvi1.Diabetes where bmi > (select avg(bmi) from guvi1.Diabetes) " )  # using sub query - fetch the count of diabetic patients whose bmi is greater than average bmi
out=mycursor.fetchall()   # fetching all the data
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))   # psql is like sql
```
