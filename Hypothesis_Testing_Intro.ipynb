```python
#Z-test
from statsmodels.stats.weightstats import ztest as ztest
import scipy.stats as stats


#enter IQ levels for 20 patients
data = [88, 92, 94, 94, 96, 97, 97, 97, 99, 99,105, 109, 109, 109, 110, 112, 112, 113, 114, 115]


#perform one sample z-test
print(ztest(data, value=103.05))
print(stats.ttest_1samp(data, popmean=103.05))
#value is nothing but H0 or Null hypothesis(here we say that avg is 100)
```

```python
import pandas as pd
df = pd.read_csv("/content/diabetes.csv")
display(df.head())
```

```python
#BMI - Z-Test
data = df['BMI']
#perform one sample z-test
print(ztest(data, value=32))
print(stats.ttest_1samp(data, popmean=32))
```

```python
#BMI - Z-Test
data = df['Pregnancies']
#perform one sample z-test
print(ztest(data, value=4))
print(stats.ttest_1samp(data, popmean=4))
```
