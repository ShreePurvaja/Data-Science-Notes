# **Extract month name from date and store in new column named as month**

```python
df['Month']=df.index.month
df['Month']=df['Month'].apply(lambda x:calendar.month_name[x])
df.head()
```

# **List of month in which rained with thunderstorm**

```python
def checkList(x):
  weatherlist=['rain','thunderstorm']
  count=0
  for substring in weatherlist:
    if substring in x:
      count+=1
  if(count==2):
    return True
  else:
    return False

print("List of month in which rained with thunderstorm")
rained=df['Weather'].str.lower().apply(checkList)
month_list=df[rained]['Month'].unique()
print(month_list)
```
