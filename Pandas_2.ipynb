```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/content/tips.csv")
df
```

```python
df.groupby(df['day']).count()
```

# **Count how many are from female**

```python
df.groupby(['sex']).count()
```

# **Resetting the index**

```python
df.groupby('day')['tip'].sum().reset_index()
```

# **Sorting the values by tips**

```python
df.groupby('day')['tip'].sum().reset_index().sort_values(by='tip')
```

# **Pivot Table**

```python
df=pd.read_csv("/content/CAR DETAILS FROM CAR DEKHO.csv")
df
```

```python
df.head()
```

# **Pass multiple indices**

```python
pd.pivot_table(df, index=['name','fuel'], values=['year','selling_price', 'km_driven'])  # Example numeric columns
```

```python
pd.pivot_table(df, index=['fuel'], values=['year','selling_price', 'km_driven'])  # Example numeric columns
```

```python
pd.pivot_table(df, index=['transmission'], values=['year','selling_price', 'km_driven'])  # Example numeric columns
```

```python
pd.pivot_table(df, index=['seller_type'], values=['year','selling_price', 'km_driven'])  # Example numeric columns
```

```python
pd.pivot_table(df, index=['owner'], values=['year','selling_price', 'km_driven'])  # Example numeric columns
```

```python
pd.pivot_table(df, index=['fuel','owner','transmission'], values=['year','selling_price', 'km_driven'])  # Example numeric columns
```

```python
pd.pivot_table(df, index=['fuel','transmission'], values=['year','selling_price', 'km_driven'],aggfunc=[len,np.mean])
```

```python
pd.pivot_table(df, index=['fuel','transmission'], values=[ 'km_driven'],aggfunc=[np.sum])
```
