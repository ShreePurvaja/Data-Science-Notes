# **Data Collection**

```python
import pandas as pd
data=pd.read_csv("https://raw.githubusercontent.com/nethajinirmal13/Training-datasets/refs/heads/main/insurance.csv")
data
```

# **Split the data**

```python
x=data[['age','sex','smoker','region']]
y=data['charges']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

# **Encoding the label**

```python
# prompt: x=data[['age','sex','smoker']]
# create a lable encoding for the above code

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
x['sex'] = le.fit_transform(x['sex'])
x['region'] = le.fit_transform(x['region'])
x['smoker'] = le.fit_transform(x['smoker'])

print(x.head())
```

```python
#prompt : I dont't want 0 in encoding
from sklearn.preprocessing import LabelEncoder

#create a label encoder object
le = LabelEncoder()

#Avoid 0 in encoding by adding 1 to the encoded values
#fit and transform the columns need to be encoded
x['sex_encoded'] = le.fit_transform(x['sex']) + 1
x['region_encoded'] = le.fit_transform(x['region']) + 1
x['smoker_encoded'] = le.fit_transform(x['smoker']) + 1


#drop the original values
x=x.drop(['sex','smoker','region'],axis=1)

#now the columns have numeric values
x
```

# **Set Features and Target for modified data**

```python
# Select features and target
x = x[['age', 'sex_encoded', 'smoker_encoded', 'region_encoded']]
y = data['charges']
```

Split data

```python
# 70% train, 30% test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

# **Preprocessing and scalling the data**

```python
# Initialize scaler
scaler = MinMaxScaler()

# Fit and transform training data
x_train = scaler.fit_transform(x_train)

# Show min, max, range
print("Max values:", scaler.data_max_)
print("Min values:", scaler.data_min_)
print("Ranges:", scaler.data_range_)

# Transform test data
x_test = scaler.transform(x_test)
```

# **Training the model**

```python
model = LinearRegression()
model.fit(x_train, y_train)

print("Model Coefficients:", model.coef_)
print("Model Intercept:", model.intercept_)
```

# **Testing the model**

```python
# Example new data
new = pd.DataFrame({
    'age': [25],
    'sex_encoded': [2],
    'smoker_encoded': [2],
    'region_encoded': [3]
})

# Scale new input
new_scaled = scaler.transform(new)

# Predict
prediction = model.predict(new_scaled)
print("Predicted Charges:", prediction)
```

# **Model evaluation**

```python
from sklearn.metrics import r2_score
print(r2_score(y_test,model.predict(x_test)))
```
