```python
# mean,median,mode for bloodpressure

import pandas as pd
from scipy import stats

df = pd.read_csv("/content/diabetes.csv")

# Calculate mean, median, and mode for BloodPressure
mean_bp = df['BloodPressure'].mean()
median_bp = df['BloodPressure'].median()
mode_bp = df['BloodPressure'].mode()[0]  # Extract the mode value

print(f"Mean Blood Pressure: {mean_bp}")
print(f"Median Blood Pressure: {median_bp}")
print(f"Mode Blood Pressure: {mode_bp}")
```
